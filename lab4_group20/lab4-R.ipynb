```r
install.packages("ggrepel")
library(ggrepel)
```

```r
library(readr)
library(dplyr)
library(ggplot2)

# 1) Read data quietly (no column specification message)
languages <- read_csv(
  "/Users/bbz/Desktop/languages.csv",
  show_col_types = FALSE
)

# 2) Compute z-score within each continent and label potential outliers
languages_q1 <- languages %>%
  group_by(continent) %>%
  mutate(
    mu  = mean(information_density, na.rm = TRUE),
    sig = sd(information_density, na.rm = TRUE),
    z_score = if_else(sig > 0, (information_density - mu) / sig, 0),
    label_lang = if_else(abs(z_score) > 2, language, "")
  ) %>%
  ungroup() %>%
  select(-mu, -sig)

# 3) Plot (suppress geom_smooth message by setting it as a layer option)
p1 <- ggplot(
  languages_q1,
  aes(x = distinct_syllables, y = information_density, colour = continent)
) +
  geom_point(size = 2.2, alpha = 0.85) +
  geom_smooth(method = "lm", se = FALSE, linewidth = 0.9) +
  geom_text(
    data = languages_q1 %>% filter(label_lang != ""),
    aes(label = label_lang),
    nudge_x = 200,
    nudge_y = 0.05,
    check_overlap = TRUE,
    show.legend = FALSE
  ) +
  scale_x_continuous(expand = expansion(mult = 0.05)) +
  scale_y_continuous(expand = expansion(mult = 0.05)) +
  labs(
    title = "Information density vs. distinct syllables",
    x = "Distinct syllables",
    y = "Information density (bits per syllable)",
    colour = "Continent"
  ) +
  theme_minimal(base_size = 12) +
  theme(legend.position = "right")

# Suppress the "geom_smooth() using formula..." message
suppressMessages(print(p1))
```

# 2. Speech rate and information density

Hmm... what an intriguing relationship we just revealed... letâ€™s keep these clusters in mind while exploring whether continuing our exploration. The two variables that directly determine the information rate of a language are its information density (bits of information per syllable) and the rate at which it is spoken (syllables per second). A high value in both would indicate a high information rate and efficient communication (a higher number of information bits conveyed per second).

```r
library(readr)
library(dplyr)
library(ggplot2)
library(ggrepel)

# 0) Read the spoken-texts dataset (make sure the object name matches)
spoken_texts <- read_csv(
  "/Users/bbz/Desktop/spoken-texts.csv",
  show_col_types = FALSE
)

# 1) Compute average speech rate (syllables per second) per language
spoken_rates <- spoken_texts %>%
  mutate(speech_rate = syllables / duration) %>%
  group_by(iso_lang) %>%
  summarise(
    speech_rate = mean(speech_rate, na.rm = TRUE),
    .groups = "drop"
  )

# 2) Join speech rate onto the language-level table from Q1
#    (languages_q1 must already exist from Q1)
languages_q2 <- languages_q1 %>%
  left_join(spoken_rates, by = "iso_lang") %>%
  filter(!is.na(speech_rate))

# 3) Plot: information density vs speech rate
p2 <- ggplot(
  languages_q2,
  aes(x = information_density, y = speech_rate, colour = continent)
) +
  geom_point(size = 2.2, alpha = 0.85) +
  geom_smooth(method = "lm", se = FALSE, linewidth = 0.9) +
  geom_text_repel(
    data = languages_q2 %>% filter(label_lang != ""),
    aes(label = label_lang),
    min.segment.length = 0,
    show.legend = FALSE
  ) +
  scale_x_continuous(expand = expansion(mult = 0.05)) +
  scale_y_continuous(expand = expansion(mult = 0.05)) +
  labs(
    x = "Information density (bits per syllable)",
    y = "Speech rate (syllables per second)",
    colour = "Continent"
  ) +
  theme_minimal()

p2
```

### Question 3

<h4>R</h4>
    
<ol type="1">
<li>It looks like we do not have a language that is high both in information density and speech rate, which could indicate that human minds are not good at processing auditory information beyond a certain rate limit (read more in the article if you're interested). Letâ€™s directly plot the rate at which information is conveyed for all the languages to see if there are at least small differences between them. Combine the appropriate columns to create a new column, <code>information_rate</code>, which holds the amount of information per second conveyed by each speaker.</li>
<li>Create a plot that has the language ISO codes on one axis and the information rate on the other. Plot a graphical representation of either the distribution of data or every single data point. Color this by the same variable as in the previous two scatter plots. Sort the languages in a suitable order.</li>

```r
library(dplyr)
library(ggplot2)

# ------------------------------------------------------------
# 1) Build record-level dataset with information_rate
# ------------------------------------------------------------
df3 <- spoken_texts %>%
  mutate(speech_rate = syllables / duration) %>%
  left_join(
    languages %>% select(iso_lang, language, information_density, continent),
    by = "iso_lang"
  ) %>%
  mutate(information_rate = speech_rate * information_density) %>%
  # Keep only rows we can actually plot
  filter(!is.na(information_rate), !is.na(continent), !is.na(language))

# ------------------------------------------------------------
# 2) Summarize mean information_rate + 95% CI by language
#    CI = mean Â± 1.96 * SE, SE = sd / sqrt(n)
#    Handle n=1 safely (sd is NA -> set SE = 0)
# ------------------------------------------------------------
summary3 <- df3 %>%
  group_by(iso_lang, language, continent) %>%
  summarise(
    mean_rate = mean(information_rate, na.rm = TRUE),
    sd_rate   = sd(information_rate, na.rm = TRUE),
    n         = n(),
    .groups   = "drop"
  ) %>%
  mutate(
    se = if_else(n > 1, sd_rate / sqrt(n), 0),
    ci_lower = mean_rate - 1.96 * se,
    ci_upper = mean_rate + 1.96 * se
  )

# Order ISO codes by mean_rate (for x-axis ordering)
order_iso <- summary3 %>%
  arrange(mean_rate) %>%
  pull(iso_lang)

# Create a single ordered factor column to reuse in all layers
df3 <- df3 %>% mutate(iso_lang_f = factor(iso_lang, levels = order_iso))
summary3 <- summary3 %>% mutate(iso_lang_f = factor(iso_lang, levels = order_iso))

# ------------------------------------------------------------
# 3) Plot: raw points (jitter) + 95% CI error bars + mean points
# ------------------------------------------------------------
p3 <- ggplot() +
  # Raw observations
  geom_point(
    data = df3,
    aes(x = iso_lang_f, y = information_rate, colour = continent),
    alpha = 0.40,
    size = 1.6,
    position = position_jitter(width = 0.15, height = 0)
  ) +
  # 95% CI for the mean
  geom_errorbar(
    data = summary3,
    aes(x = iso_lang_f, ymin = ci_lower, ymax = ci_upper),
    width = 0.12,
    linewidth = 0.7
  ) +
  # Mean point
  geom_point(
    data = summary3,
    aes(x = iso_lang_f, y = mean_rate),
    size = 2.3,
    colour = "black"
  ) +
  labs(
    x = "ISO language code",
    y = "Information rate (bits per second)",
    colour = "Continent"
  ) +
  scale_y_continuous(expand = expansion(mult = 0.05)) +
  theme_minimal(base_size = 12) +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1)  # rotate labels if crowded
  )

p3
```

Plot A â€” Distinct syllables vs. information density

Figure A compares each languageâ€™s information density (bits per syllable) with its number of distinct syllables. The points cluster rather than forming one smooth cloud, and coloring by continent suggests that part of this clustering aligns with geographic/linguistic regions. In other words, languages differ in both how large their syllable inventory is and how much information each syllable tends to carry, and those differences are not randomly distributed. This motivates the next question: if languages differ in information per syllable, do speakers compensate in how fast they speak?

Plot B â€” Information density vs. speech rate

Figure B shows a compensation pattern between information density and speech rate (syllables per second). Languages with higher information density tend to be spoken at a lower syllable rate, while languages with lower information density tend to be spoken faster. The continent-specific trend lines broadly support the same idea, even if the exact slope differs slightly by region. This suggests that languages may reach similar communicative efficiency through different strategiesâ€”either packing more information into each syllable or producing syllables more quickly. That naturally leads to the final step: if these two components trade off, what happens to the overall information transmitted per second?

Plot C â€” Information rate per language

Figure C combines the two components into a single measure: information rate (bits per second), computed as (syllables/second) Ã— (bits/syllable). Despite the clear variation in Figure A and the trade-off in Figure B, the average information rate by language falls into a relatively narrow range, with only a few languages standing out as higher or lower (as reflected by the confidence intervals). The key takeaway is that languages can look very different in their sound systems and timing, but they often converge toward similar overall information throughput via compensation between information density and speech rate.
